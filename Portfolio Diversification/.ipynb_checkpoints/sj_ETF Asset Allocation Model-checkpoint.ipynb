## ETF Asset Allocation Model

```python
%matplotlib inline
import zipline
from zipline.api import order_target_percent, symbol, schedule_function, date_rules, time_rules
from datetime import datetime
import pytz
from matplotlib import pyplot as plt
import pandas as pd
import pyfolio as pf
from trading_calendars import get_calendar
```

```python
def initialize(context):
    # Securities and target weights
    context.securities = {'AAPL': 0.5,'MSFT': 0.5}
# {500209 : 0.5, 500470 : 0.5}
#                         'AAPL': 0.5, #0.25, 
#                         'MSFT': 0.5
#                         'ABBV': 0.3, 
#                         'ABC': 0.3, 
#                         'CAT': 0.075, 
#                         'CSCO': 0.075
#                         'INFY.NS': 0.25, 
#                         'LT.NS': 0.3, 
#                         'HDFC.NS': 0.3, 
#                         'HDFCBANK.NS': 0.075, 
#                         'HCLTECH.NS': 0.075
#     }
# AAPL","AXP","BA","CAT","CSCO"
    # Schedule rebalance for once a month
    schedule_function(rebalance, date_rules.month_start(), time_rules.market_open())
```

```python
def rebalance(context, data):
    # Loop through the securities
    for sec, weight in context.securities.items():
        sym = symbol(sec)
        # Check if we can trade
        if data.can_trade(sym):
            # Reset the weight
            order_target_percent(sym, weight)    
```

```python
def analyze(context, perf):
    # Use PyFolio to generate a performance report
    returns, positions, transactions = pf.utils.extract_rets_pos_txn_from_zipline(perf)
    pf.create_returns_tear_sheet(returns, benchmark_rets=None)
```

```python
# Set start and end date
# start = datetime(2016, 1, 1, 8, 15, 12, 0, pytz.UTC)
# end = datetime(2018, 12, 31, 8, 15, 12, 0, pytz.UTC)

start = pd.Timestamp('2016-01-01',).tz_localize('utc') #.tz_localize('US/Eastern')   #.tz_localize('utc') #('US/Eastern') #tz=UTC)
end = pd.Timestamp('2020-01-01', ).tz_localize('utc') #.tz_localize('US/Eastern')   #.tz_localize('utc') #('US/Eastern') #tz=UTC)
```

```python
# Fire off backtest
result = zipline.run_algorithm(
    start=start, # Set start
    end=end,  # Set end
    analyze=analyze, 
    initialize=initialize, # Define startup function
    capital_base=100000, # Set initial capital
    data_frequency = 'daily',  # Set data frequency
    bundle= 'quandl') 
#'random_equities') #'india_stock_data' )#'quandl') #'ac_equities_db' csvdir) # Select bundle bse_data
```
